```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/insurance.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print("Dataset Preview:")
print(data.head())
```

```python

### a) Create vectors, factors, lists, matrices, and data frames ###
# Vectors
age_vector = data['age']
smoker_vector = data['smoker']

# Factors (Categorical data)
sex_factor = pd.Categorical(data['sex'])

# List
custom_list = [age_vector.mean(), data['expenses'].mean(), "Insurance Data Analysis"]

# Matrix (example: first 10 rows of 'age' and 'expenses')
matrix = np.array([data['age'][:10], data['expenses'][:10]])

# Data frame (subset of the original dataset)
subset_df = data[['age', 'smoker', 'expenses']]

# Print created data structures
print("\nVector (age):")
print(age_vector[:5])
print("\nFactor (sex):")
print(sex_factor[:5])
print("\nList:")
print(custom_list)
print("\nMatrix:")
print(matrix)
print("\nSubset DataFrame:")
print(subset_df.head())

```

```python
### b) Scatter plot between age and smoking ###
plt.figure(figsize=(10, 6))
plt.scatter(data['age'], data['smoker'].apply(lambda x: 1 if x == 'yes' else 0), c='blue', marker='+', alpha=0.7)
plt.title("Age vs Smoking Status", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Smoking Status (1: Smoker, 0: Non-smoker)", fontsize=12)
plt.grid(True)
plt.show()


```

```python
### Scatter plot between age (20 to 40) and smoking ###
filtered_age_data = data[(data['age'] >= 20) & (data['age'] <= 40)]
plt.figure(figsize=(10, 6))
plt.scatter(filtered_age_data['age'], filtered_age_data['smoker'].apply(lambda x: 1 if x == 'yes' else 0), c='green', marker='o', alpha=0.7)
plt.title("Age (20-40) vs Smoking Status", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Smoking Status (1: Smoker, 0: Non-smoker)", fontsize=12)
plt.grid(True)
plt.show()
```

```python
### Find expenses of smokers and non-smokers ###
expenses_by_smoker = data.groupby('smoker')['expenses'].mean()
expenses_by_smoker.plot(kind='bar', color=['orange', 'blue'], figsize=(8, 6))
plt.title("Average Expenses: Smokers vs Non-Smokers", fontsize=16)
plt.ylabel("Average Expenses", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

```

```python
### Find the number of smokers gender-wise ###
smokers_gender = data[data['smoker'] == 'yes'].groupby('sex').size()
smokers_gender.plot(kind='bar', color=['purple', 'cyan'], figsize=(8, 6))
plt.title("Number of Smokers by Gender", fontsize=16)
plt.ylabel("Number of Smokers", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

```

```python
### Scatter plot between age and expenses ###
plt.figure(figsize=(10, 6))
plt.scatter(data['age'], data['expenses'], c='red', alpha=0.6, edgecolors='k')
plt.title("Age vs Expenses Scatter Plot", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Expenses", fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Filter data for teenage smokers (ages 13 to 19) and smoking status as 'yes'
teenage_smokers = data[(data['age'] >= 13) & (data['age'] <= 19) & (data['smoker'] == 'yes')]

# Count the number of teenage smokers
teenage_smokers_count = teenage_smokers.shape[0]
print(f"Number of Teenage Smokers: {teenage_smokers_count}")

# Plot the count of teenage smokers
plt.figure(figsize=(6, 4))
plt.bar(['Teenage Smokers'], [teenage_smokers_count], color='skyblue')
plt.title("Count of Teenage Smokers", fontsize=16)
plt.ylabel("Count", fontsize=12)
plt.grid(axis='y')
plt.show()
```
